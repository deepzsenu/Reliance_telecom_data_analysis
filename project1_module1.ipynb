
# Query 1.5 Write a function to calculate the average price over the last N days of the stock price data

##### where N is a user defined parameter.
##### Write a second function to calculate the profit/loss percentage over the last N days.
##### Calculate the average price AND the profit/loss percentages over the course of last

    1 week,
    2 weeks,
    1 month,
    3 months,
    6 months and
    1 year.

{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }


```python
def avgNDayPrice(N): #function to check the average of the last n weeaks
    return df['Average Price'].tail(N).mean()
def profitLossNDays(N):
    change = (df['Close Price'].tail(N).iloc[N-1] - df['Close Price'].tail(N).iloc[0])
    profit_loss_percen = (change/df['Close Price'].tail(N).iloc[N-1])*100
    return abs(profit_loss_percen)
```

```python
print("Average prices for last N days are as follows:")

days = int(input("Enter the last N days upto which you want the average price : "))
print("Last {} Days average price :".format(days),round(avgNDayPrice(days),2))
```

```python
print("Loss/Profit percentage for last N days are as follows:")

p_l_days = int(input("Enter the last N days upto which you want the Loss/Profit percentage : "))

print("Last {} Datys Loss/Profit percentage is: ".format(p_l_days),round(profitLossNDays(p_l_days),2))
```


# Query 1.7 Add another column 'Trend'

- ### whose values are:

    - 'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
    - 'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
    - 'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
    - 'Positive' for 'Day_Perc_Change' in between 1 and 3
    - 'Negative' for 'Day_Perc_Change' in between -1 and -3
    - 'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
    - 'Among top losers' for 'Day_Perc_Change' in between -3 and -7
    - 'Bull run' for 'Day_Perc_Change' >7
    - 'Bear drop' for 'Day_Perc_Change' <-7



```python
df['Trend'] = np.nan
for i in range(len(df)):
    #'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
    if ((df['Day_Perc_Change'][i] >= -0.5) & (df['Day_Perc_Change'][i] <= 0.5)):
    
        df['Trend'][i] = 'Slight or No change'
    
    #'slight positive' for 'Day_Perc_Change' in between 0.5 and 1
    if ((df['Day_Perc_Change'][i] >= 0.5) & (df['Day_Perc_Change'][i] <= 1)):
    
        df['Trend'][i] = 'Slight positive'
    
    #'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
    if ((df['Day_Perc_Change'][i] <= -0.5) & (df['Day_Perc_Change'][i] >= -1)):
    
        df['Trend'][i] = 'Slight negative'
    
    #'Positive' for 'Day_Perc_Change' in between 1 and 3
    if ((df['Day_Perc_Change'][i] >= 1) & (df['Day_Perc_Change'][i] <= 3)):
    
        df['Trend'][i] = 'Positive' 
    
    #'Negative' for 'Day_Perc_Change' in between -1 and -3
    if ((df['Day_Perc_Change'][i] <= -1) & (df['Day_Perc_Change'][i] >= -3)):
    
        df['Trend'][i] = 'Negative'
    
    #'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
    if ((df['Day_Perc_Change'][i] >= 3) & (df['Day_Perc_Change'][i] <= 7)):
    
        df['Trend'][i] = 'Among top gainers'
    
    #'Among top losers' for 'Day_Perc_Change' in between -3 and -7   
    if ((df['Day_Perc_Change'][i] <= -3) & (df['Day_Perc_Change'][i] >= -7)):
    
        df['Trend'][i] = 'Among top losers'
    
    #'Bull run' for 'Day_Perc_Change' >7
    if (df['Day_Perc_Change'][i] > 7):
    
        df['Trend'][i] = 'Bull run' 
    
    #'Bear drop' for 'Day_Perc_Change' <-7
    if (df['Day_Perc_Change'][i] < -7):
    
        df['Trend'][i] = 'Bear drop' 
    

    
```
